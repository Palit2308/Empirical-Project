```python
import pandas as pd 
```

```python
from sklearn.preprocessing import StandardScaler

cols_for_standardization = df1.columns[2:]  # standardizing the columns for better results

scaler = StandardScaler()

standardized_data = scaler.fit_transform(df1[cols_for_standardization])

standardized_df = pd.DataFrame(standardized_data, columns=cols_for_standardization)

non_standardized_df = df1[['Country', 'Country Code']]

df1 = pd.concat([non_standardized_df, standardized_df], axis=1)  # to get the final dataframe after standardization
```

```python
df1.head()
```

```python
# performing KNN MEans imputation for the missing values with 5 neighbours


from sklearn.impute import KNNImputer

cols_for_imputation = df1.columns[2:]  
knn_imputer = KNNImputer(n_neighbors=5)

df1[cols_for_imputation] = knn_imputer.fit_transform(df1[cols_for_imputation])
```

```python
df1.head()
```

```python
# to perform hierarchical clustering 4 times and to drop the countries belonging to clusters with less than 4 countries

from scipy.cluster.hierarchy import linkage, fcluster
from collections import Counter
import numpy as np

np.random.seed(42)

df2 = pd.DataFrame(columns=df1.columns)   # to keep the dropped countries in this new empty dataframe

num_iterations = 4

for i in range(num_iterations):

    cols_for_clustering = df1.columns[2:] 
   
    linkage_matrix = linkage(df1[cols_for_clustering], metric='mahalanobis')   # Perform hierarchical clustering with mahalanobis distance
 
    outlier_threshold = 5   # defining a threshold for identifying outliers

    clusters = fcluster(linkage_matrix, t=outlier_threshold, criterion='distance')

    df1_copy = df1.copy()

    df1_copy['Cluster'] = clusters     # Now, 'df1' contains the cluster assignments for each country, taking into account the weights

    cluster_counts = Counter(df1_copy['Cluster'])   # Count the number of countries in each cluster

   
    small_clusters = [cluster for cluster, count in cluster_counts.items() if count < 4]    # Identifying clusters with fewer than 4 countries

    for cluster in small_clusters:
        country_codes_to_drop = df1_copy[df1_copy['Cluster'] == cluster]['Country Code'].tolist()
        print(f"Iteration {i+1}: Country Codes in Cluster {cluster} to be dropped: {', '.join(country_codes_to_drop)}")

        df2 = pd.concat([df2, df1_copy[df1_copy['Country Code'].isin(country_codes_to_drop)]], ignore_index=True)

   
    df1_copy = df1_copy[~df1_copy['Cluster'].isin(small_clusters)]    # Dropping countries belonging to small clusters

    df1_copy.drop(columns=['Cluster'], inplace=True)

    df1 = df1_copy

```

```python
# performimg the KNN Means to identify 5 clusters after dropping the outliers

from sklearn.cluster import KMeans
import pandas as pd


np.random.seed(42)

numeric_cols = df1.drop(columns=['Country', 'Country Code'])

kmeans = KMeans(n_clusters=5)

kmeans.fit(numeric_cols)

cluster_labels = kmeans.labels_  # Get cluster labels for each data point

df1['Cluster'] = cluster_labels  # Adding cluster labels to the DataFrame



```
